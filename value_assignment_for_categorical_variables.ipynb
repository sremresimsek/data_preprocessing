```python
import numpy as np
import pandas as pd
```

```python
V1 = np.array([1, 3, 6, np.nan, 7, 1, np.nan, 9, 15])
V4 = np.array(["IT", np.nan, "IK", "IK", "IK", "IK", "IK", "IT", "IT"], dtype = object)

df = pd.DataFrame({
    "salary" : V1,
    "department" : V4
})

df
```

```python
df["department"].mode()
```

```python
df["department"].fillna(df["department"].mode()[0])
```

```python
df["department"].fillna(method = "bfill") # Filling with next value
```

```python
df.department.fillna(method = "ffill") # Filling with previous value
```
