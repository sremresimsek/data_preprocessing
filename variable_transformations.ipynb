```python
import numpy as np
import pandas as pd
import seaborn as sns
df = sns.load_dataset("tips")
df.head()
```

## 0-1 Transformation

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lbe = LabelEncoder()
```

```python
lbe.fit_transform(df["sex"])
```

```python
df["new_sex"] = lbe.fit_transform(df["sex"])
```
