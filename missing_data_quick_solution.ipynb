## Deleting Methods

```python
V1 = np.array([1, 3, 6, np.NaN, 7, 1, np.NaN, 9, 15])
V2 = np.array([7, np.NaN, 5, 8, 12, np.NaN, np.NaN, 2, 3])
V3 = np.array([np.NaN, 12, 5, 6, 14, 7, np.NaN, 2, 3])
df = pd.DataFrame({"V1" : V1, "V2" : V2, "V3" : V3})
df
```

```python
df.dropna()
```

```python
df.dropna(how = "all")
```

```python
df.dropna(axis = 1)
```

```python
df["sil_beni"] = np.nan
```

```python
df
```

```python
df.dropna(axis = 1, how = "all")
```

```python
df.dropna(axis = 1, how = "all", inplace = True)
```

## Simple Value Assignment Methods

```python
df
```

```python
df.V1.fillna(1)
```

```python
df["V1"].mean()
```

```python
df["V1"].fillna(df["V1"].mean())
```

```python
df.apply(lambda x: x.fillna(x.mean()), axis = 1) # filling with row-based averaging
```

```python
df.apply(lambda x: x.fillna(x.mean()), axis = 0) # filling with column-based averaging
```

```python
df.fillna(df.mean()[:])
```

```python
df.fillna(df.mean()["V1": "V2"])
```

```python
df.V3.fillna(df.V3.median())
```

```python
df.where(pd.notna(df), df.mean(), axis = "columns")
```
