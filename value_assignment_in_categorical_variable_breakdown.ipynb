```python
import numpy as np
import pandas as pd
```

```python
V1 = np.array([1, 3, 6, np.nan, 7, 1, np.nan, 9, 15])
V2 = np.array([7, np.nan, 5, 8, 12, np.nan, np.nan, 2, 3])
V3 = np.array([np.nan, 12, 5, 6, 14, 7, np.nan, 2, 31])
V4 = np.array(["IT", "IT", "IK", "IK", "IK", "IK", "IK", "IT", "IT"])

df = pd.DataFrame({
    "salary" : V1,
    "V2" : V2,
    "V3" : V3,
    "department" : V4
})

df
```

```python
df.groupby("department")["salary"].mean()
```
